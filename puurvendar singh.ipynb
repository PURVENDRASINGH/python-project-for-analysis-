```python
grouped_df = df.groupby('State')
print(grouped_df)
```

```python
# Group by 'State' and calculate the sum of other numeric columns
grouped_df = df.groupby('State').count()

# Display the grouped data
print(grouped_df)
```

```python
# Group by 'State' and calculate the mean of other numeric columns
grouped_df = df.groupby('State').sum()

# Display the grouped data
print(grouped_df)
```

```python

x = grouped_df['Amount'].index

y = grouped_df['Amount'].values
print(x,y)
```

```python


# Assuming 'x' contains the State names and 'y' contains the summed Amounts
plt.figure(figsize=(10, 6))

# Create a bar chart
plt.bar(x, y)

# Add data labels on top of each bar
for i, v in enumerate(y):
    plt.text(i, v + 0.1, str(v), ha='center', va='bottom')  # Positioning the text above the bars

# Add labels and title
plt.xlabel('State')
plt.ylabel('Total Amount')
plt.title('Total Amount by State')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Automatically adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```
